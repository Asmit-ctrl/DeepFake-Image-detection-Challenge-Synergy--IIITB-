```python
# Cell 1: Imports and Configuration
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from transformers import ViTModel
from torchvision import transforms
from PIL import Image
import os
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import seaborn as sns
import datetime

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
```

```python

# Cell 2: Improved Focal Loss
class ImprovedFocalLoss(nn.Module):
    def __init__(self, alpha=0.25, gamma=2.0, smoothing=0.1):
        super(ImprovedFocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.smoothing = smoothing
        
    def forward(self, inputs, targets):
        BCE_loss = nn.functional.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-BCE_loss)
        F_loss = self.alpha * (1-pt)**self.gamma * BCE_loss
        return F_loss.mean()
```

```python



# Cell 3: ViT-Base Model 
class PureViTBase(nn.Module):
    def __init__(self, num_classes=2):
        super(PureViTBase, self).__init__()
        
        print("ğŸš€ Loading Pure ViT-Base (86M parameters)...")
        
        # Load pre-trained ViT-Base
        self.vit = ViTModel.from_pretrained('google/vit-base-patch16-224')
        hidden_size = self.vit.config.hidden_size  # 768 for ViT-Base
        
        # Efficient Classifier optimized for Base model
        self.classifier = nn.Sequential(
            nn.Linear(hidden_size, 384),
            nn.BatchNorm1d(384),
            nn.GELU(),
            nn.Dropout(0.3),
            
            nn.Linear(384, 192),
            nn.BatchNorm1d(192),
            nn.GELU(),
            nn.Dropout(0.2),
            
            nn.Linear(192, num_classes)
        )
        
        # Initialize classifier weights
        self._initialize_weights()
        
        print(f"âœ… Pure ViT-Base loaded successfully!")
        print(f"ğŸ“Š Total parameters: {sum(p.numel() for p in self.parameters()):,}")
        print(f"ğŸ“Š Model size: ~330MB (runs on 8GB+ GPUs)")
    
    def _initialize_weights(self):
        for m in self.classifier.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
    
    def forward(self, pixel_values):
        outputs = self.vit(pixel_values)
        cls_token = outputs.last_hidden_state[:, 0, :]
        logits = self.classifier(cls_token)
        return logits
```

```python
# Cell 4: Data Transforms
def get_vit_transforms(augment=False):
    if augment:
        return transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.RandomCrop((224, 224)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomRotation(degrees=10),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
        ])
    else:
        return transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
        ])
```

```python

# Cell 5: Updated Dataset Class for Separate Folders
class SeparateFolderDataset(Dataset):
    def __init__(self, real_dir, fake_dir, transform=None):
        self.image_paths = []
        self.labels = []
        self.transform = transform
        
        print("ğŸ“ Loading dataset...")
        
        # Load real images
        real_count = 0
        if os.path.exists(real_dir):
            for img_name in os.listdir(real_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
                    self.image_paths.append(os.path.join(real_dir, img_name))
                    self.labels.append(0)
                    real_count += 1
            print(f"âœ… Found {real_count} real images in {real_dir}")
        else:
            print(f"âŒ Real directory not found: {real_dir}")
        
        # Load fake images
        fake_count = 0
        if os.path.exists(fake_dir):
            for img_name in os.listdir(fake_dir):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
                    self.image_paths.append(os.path.join(fake_dir, img_name))
                    self.labels.append(1)
                    fake_count += 1
            print(f"âœ… Found {fake_count} fake images in {fake_dir}")
        else:
            print(f"âŒ Fake directory not found: {fake_dir}")
        
        if len(self.image_paths) == 0:
            raise ValueError(f"ğŸš« No images found! Check your directories.")
        
        print(f"ğŸ“Š Dataset summary: {len(self.image_paths)} total images")
        print(f"ğŸ“Š Class balance: {real_count} real, {fake_count} fake")
    
    def __len__(self):
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        try:
            image = Image.open(self.image_paths[idx]).convert('RGB')
            label = self.labels[idx]
            
            if self.transform:
                image = self.transform(image)
            
            return image, label
            
        except Exception as e:
            print(f"âŒ Error loading {self.image_paths[idx]}: {e}")
            # Return a dummy tensor with correct shape
            dummy_image = torch.randn(3, 224, 224)
            return dummy_image, 0
```

```python




# Cell 6: Trainer Optimized for ViT-Base
class ViTBaseTrainer:
    def __init__(self):
        print("ğŸš€ Initializing ViT-Base Trainer...")
        self.model = PureViTBase()
        self.model.to(device)
        print("âœ… ViT-Base trainer ready!")
    
    def train(self, train_loader, val_loader, num_epochs=20, learning_rate=3e-4):
        # Separate parameters for different learning rates
        vit_params = []
        classifier_params = []
        
        for name, param in self.model.named_parameters():
            if 'classifier' in name:
                classifier_params.append(param)
            else:
                vit_params.append(param)
        
        # Optimizer with adjusted learning rates for Base model
        optimizer = optim.AdamW([
            {'params': vit_params, 'lr': learning_rate / 5, 'weight_decay': 0.01},
            {'params': classifier_params, 'lr': learning_rate, 'weight_decay': 0.02}
        ], betas=(0.9, 0.999))
        
        # Cosine Annealing scheduler
        scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
            optimizer, T_0=8, T_mult=1, eta_min=1e-6
        )
        
        criterion = ImprovedFocalLoss(alpha=0.25, gamma=2.0)
        
        train_losses, val_accuracies, val_losses = [], [], []
        best_accuracy = 0.0
        patience = 10
        patience_counter = 0
        
        print(f"ğŸ”¥ Training ViT-Base for {num_epochs} epochs...")
        print(f"ğŸ“Š Learning Rate Config: Backbone={learning_rate/5:.2e}, Classifier={learning_rate:.2e}")
        
        for epoch in range(num_epochs):
            # Training phase
            self.model.train()
            running_loss = 0.0
            batch_count = 0
            
            train_pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
            for images, labels in train_pbar:
                if images.nelement() == 0:
                    continue
                    
                images, labels = images.to(device), labels.to(device)
                
                optimizer.zero_grad()
                logits = self.model(images)
                loss = criterion(logits, labels)
                loss.backward()
                
                # Gradient clipping for stability
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
                optimizer.step()
                
                running_loss += loss.item()
                batch_count += 1
                
                train_pbar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'Avg Loss': f'{running_loss/batch_count:.4f}',
                    'LR': f'{scheduler.get_last_lr()[0]:.2e}'
                })
            
            scheduler.step()
            
            # Validation phase
            val_accuracy, val_loss = self._validate(val_loader, criterion)
            
            if batch_count > 0:
                avg_train_loss = running_loss / batch_count
            else:
                avg_train_loss = 0.0
                
            train_losses.append(avg_train_loss)
            val_accuracies.append(val_accuracy)
            val_losses.append(val_loss)
            
            print(f'\nğŸ“Š Epoch [{epoch+1}/{num_epochs}] Summary:')
            print(f'   Train Loss: {avg_train_loss:.4f}')
            print(f'   Val Loss: {val_loss:.4f}')
            print(f'   Val Accuracy: {val_accuracy:.4f}')
            print(f'   Learning Rate: {scheduler.get_last_lr()[0]:.2e}')
            
            # Early stopping and model saving
            if val_accuracy > best_accuracy:
                best_accuracy = val_accuracy
                patience_counter = 0
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'val_accuracy': val_accuracy,
                    'train_loss': avg_train_loss,
                }, 'vit_base_best.pth')
                print(f'ğŸ’¾ Best ViT-Base saved! Accuracy: {best_accuracy:.4f}')
            else:
                patience_counter += 1
                if patience_counter >= patience:
                    print(f"ğŸ›‘ Early stopping triggered after {epoch+1} epochs")
                    break
        
        print(f"\nğŸ¯ Training completed! Best validation accuracy: {best_accuracy:.4f}")
        return train_losses, val_accuracies, val_losses
    
    def _validate(self, val_loader, criterion):
        self.model.eval()
        val_predictions, val_true = [], []
        val_loss = 0.0
        val_batches = 0
        
        with torch.no_grad():
            for images, labels in val_loader:
                if images.nelement() == 0:
                    continue
                    
                images, labels = images.to(device), labels.to(device)
                logits = self.model(images)
                loss = criterion(logits, labels)
                val_loss += loss.item()
                val_batches += 1
                
                _, predicted = torch.max(logits, 1)
                val_predictions.extend(predicted.cpu().numpy())
                val_true.extend(labels.cpu().numpy())
        
        if len(val_true) > 0 and val_batches > 0:
            val_accuracy = accuracy_score(val_true, val_predictions)
            avg_val_loss = val_loss / val_batches
        else:
            val_accuracy = 0.0
            avg_val_loss = 0.0
        
        return val_accuracy, avg_val_loss
```

```python

# Cell 7: Training Function for ViT-Base
def train_vit_base_separate_folders(real_train_dir, fake_train_dir, real_val_dir, fake_val_dir, num_epochs=20):
    print("ğŸš€ STARTING ViT-BASE TRAINING")
    print("=" * 70)
    print(f"ğŸ“ Training - Real: {real_train_dir}")
    print(f"ğŸ“ Training - Fake: {fake_train_dir}")
    print(f"ğŸ“ Validation - Real: {real_val_dir}")
    print(f"ğŸ“ Validation - Fake: {fake_val_dir}")
    print("=" * 70)
    
    try:
        # Create datasets
        print("\nğŸ“Š Creating datasets...")
        train_transform = get_vit_transforms(augment=True)
        val_transform = get_vit_transforms(augment=False)
        
        train_dataset = SeparateFolderDataset(real_train_dir, fake_train_dir, transform=train_transform)
        val_dataset = SeparateFolderDataset(real_val_dir, fake_val_dir, transform=val_transform)
        
        print(f"\nğŸ“Š Dataset Summary:")
        print(f"  Training images: {len(train_dataset)}")
        print(f"  Validation images: {len(val_dataset)}")
        
        if len(train_dataset) == 0:
            raise ValueError("âŒ No training images found!")
        if len(val_dataset) == 0:
            raise ValueError("âŒ No validation images found!")
        
        # Larger batch size possible with Base model
        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=0)
        val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=0)
        
        # Train model
        print("\n" + "=" * 50)
        print("ğŸŒŸ TRAINING ViT-BASE (86M parameters)")
        print("=" * 50)
        
        trainer = ViTBaseTrainer()
        train_losses, val_accuracies, val_losses = trainer.train(
            train_loader, val_loader, num_epochs=num_epochs
        )
        
        # Save final model
        print("\n" + "=" * 50)
        print("ğŸ’¾ SAVING FINAL MODEL")
        print("=" * 50)
        
        timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        model_path = f'vit_base_final_{timestamp}.pth'
        torch.save(trainer.model.state_dict(), model_path)
        print(f"ğŸ’¾ Final model saved: {model_path}")
        
        # Plot results
        if train_losses and val_accuracies:
            plt.figure(figsize=(15, 5))
            
            plt.subplot(1, 3, 1)
            plt.plot(train_losses, label='Training Loss', color='blue', linewidth=2)
            plt.plot(val_losses, label='Validation Loss', color='red', linewidth=2)
            plt.xlabel('Epochs')
            plt.ylabel('Loss')
            plt.legend()
            plt.title('ViT-Base: Training & Validation Loss')
            plt.grid(True, alpha=0.3)
            
            plt.subplot(1, 3, 2)
            plt.plot(val_accuracies, label='Validation Accuracy', color='green', marker='o')
            plt.xlabel('Epochs')
            plt.ylabel('Accuracy')
            plt.legend()
            plt.title('ViT-Base: Validation Accuracy')
            plt.grid(True, alpha=0.3)
            
            plt.subplot(1, 3, 3)
            # Show final metrics
            best_acc = max(val_accuracies) if val_accuracies else 0
            final_acc = val_accuracies[-1] if val_accuracies else 0
            metrics = ['Best Acc', 'Final Acc']
            values = [best_acc, final_acc]
            colors = ['blue', 'green']
            bars = plt.bar(metrics, values, color=colors, alpha=0.7)
            plt.ylim(0, 1)
            plt.ylabel('Accuracy')
            plt.title('ViT-Base: Final Results')
            
            # Add value labels on bars
            for bar, value in zip(bars, values):
                plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                        f'{value:.4f}', ha='center', va='bottom')
            
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            plt.show()
        
        # Final results
        best_accuracy = max(val_accuracies) if val_accuracies else 0
        final_accuracy = val_accuracies[-1] if val_accuracies else 0
        
        print(f"\nğŸ¯ ViT-BASE FINAL RESULTS:")
        print(f"  Best Validation Accuracy: {best_accuracy:.4f}")
        print(f"  Final Validation Accuracy: {final_accuracy:.4f}")
        print(f"  Training Images: {len(train_dataset)}")
        print(f"  Validation Images: {len(val_dataset)}")
        print(f"  Model saved: {model_path}")
        print(f"  Model Size: ~330MB (86M parameters)")
        print(f"  Hardware Requirement: 8GB+ VRAM (runs on RTX 3070, 4060, etc.)")
        
        return trainer, model_path
        
    except Exception as e:
        print(f"âŒ Training failed: {e}")
        import traceback
        traceback.print_exc()
        return None, None
```

```python



# Cell 8: Model Evaluation
def evaluate_vit_base(model, real_test_dir, fake_test_dir):
    """Evaluate the trained ViT-Base model"""
    print("\nğŸ“Š Evaluating ViT-Base on test data...")
    
    test_transform = get_vit_transforms(augment=False)
    test_dataset = SeparateFolderDataset(real_test_dir, fake_test_dir, transform=test_transform)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=0)
    
    model.eval()
    all_predictions, all_labels = [], []
    
    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc='Evaluating'):
            if images.nelement() == 0:
                continue
            images, labels = images.to(device), labels.to(device)
            logits = model(images)
            _, predicted = torch.max(logits, 1)
            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    accuracy = accuracy_score(all_labels, all_predictions)
    cm = confusion_matrix(all_labels, all_predictions)
    report = classification_report(all_labels, all_predictions, target_names=['Real', 'Fake'])
    
    print(f"ğŸ“Š ViT-Base Test Evaluation:")
    print(f"ğŸ¯ Accuracy: {accuracy:.4f}")
    print(f"ğŸ“Š Test samples: {len(test_dataset)}")
    print("\nğŸ“ˆ Confusion Matrix:")
    print(cm)
    print("\nğŸ“‹ Classification Report:")
    print(report)
    
    # Plot confusion matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Real', 'Fake'], 
                yticklabels=['Real', 'Fake'])
    plt.title('ViT-Base: Test Set Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    return accuracy, cm, report
```

```python
# Cell 9: Single Image Prediction
def predict_single_image_vit_base(model, image_path):
    """Predict a single image using ViT-Base"""
    try:
        # Load and transform image
        transform = get_vit_transforms(augment=False)
        image = Image.open(image_path).convert('RGB')
        input_tensor = transform(image).unsqueeze(0).to(device)
        
        # Predict
        model.eval()
        with torch.no_grad():
            logits = model(input_tensor)
            probabilities = torch.softmax(logits, dim=1)
            predicted_class = torch.argmax(logits, dim=1).item()
            confidence = probabilities[0][predicted_class].item()
        
        class_name = "FAKE" if predicted_class == 1 else "REAL"
        real_prob = probabilities[0][0].item()
        fake_prob = probabilities[0][1].item()
        
        # Display results
        plt.figure(figsize=(10, 4))
        
        plt.subplot(1, 2, 1)
        plt.imshow(image)
        plt.title(f'Input Image\n{os.path.basename(image_path)}')
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        classes = ['REAL', 'FAKE']
        probs = [real_prob, fake_prob]
        colors = ['green', 'red']
        bars = plt.bar(classes, probs, color=colors, alpha=0.7)
        plt.ylim(0, 1)
        plt.title(f'ViT-Base Prediction: {class_name}\nConfidence: {confidence:.4f}')
        plt.ylabel('Probability')
        
        # Add value labels on bars
        for bar, prob in zip(bars, probs):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                    f'{prob:.4f}', ha='center', va='bottom')
        
        plt.tight_layout()
        plt.show()
        
        print(f"\nğŸ” ViT-Base PREDICTION RESULTS:")
        print(f"   Prediction: {class_name}")
        print(f"   Confidence: {confidence:.4f}")
        print(f"   Real Probability: {real_prob:.4f}")
        print(f"   Fake Probability: {fake_prob:.4f}")
        
        return predicted_class, confidence, (real_prob, fake_prob)
        
    except Exception as e:
        print(f"âŒ Error predicting image: {e}")
        return None, None, None
```

```python
import torch
import torch.nn as nn
from transformers import ViTModel
from torchvision import transforms
from PIL import Image
import os
import json
import glob
import numpy as np
from tqdm.auto import tqdm

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class PureViTBase(nn.Module):
    def __init__(self, num_classes=2, dropout_rate=0.1):
        super(PureViTBase, self).__init__()
        self.vit = ViTModel.from_pretrained('google/vit-base-patch16-224')
        hidden_size = self.vit.config.hidden_size
        
        self.classifier = nn.Sequential(
            nn.Linear(hidden_size, 384),
            nn.BatchNorm1d(384),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(384, 192),
            nn.BatchNorm1d(192),
            nn.GELU(),
            nn.Dropout(dropout_rate),
            nn.Linear(192, num_classes)
        )
    
    def forward(self, pixel_values):
        outputs = self.vit(pixel_values=pixel_values)
        cls_token = outputs.last_hidden_state[:, 0, :]
        logits = self.classifier(cls_token)
        return logits

def load_trained_model(model_path):
    print(f"ğŸ“¥ Loading model from: {model_path}")
    model = PureViTBase(num_classes=2)
    checkpoint = torch.load(model_path, map_location=device)
    
    if 'model_state_dict' in checkpoint:
        state_dict = checkpoint['model_state_dict']
    elif 'state_dict' in checkpoint:
        state_dict = checkpoint['state_dict']
    else:
        state_dict = checkpoint
    
    model.load_state_dict(state_dict)
    model.to(device)
    model.eval()
    print("âœ… Model loaded successfully!")
    return model

def predict_test_images(model_path, test_dir, output_json='teamname_prediction.json'):
    print("ğŸš€ STARTING TEST PREDICTION PIPELINE")
    print("=" * 60)
    print(f"ğŸ“ Test images directory: {test_dir}")
    print(f"ğŸ¤– Model path: {model_path}")
    print(f"ğŸ“„ Output JSON: {output_json}")
    print("=" * 60)
    
    # Load model
    model = load_trained_model(model_path)
    
    # Image preprocessing
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ])
    
    # Find all image files
    image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.tiff']
    image_paths = []
    for extension in image_extensions:
        image_paths.extend(glob.glob(os.path.join(test_dir, extension)))
        image_paths.extend(glob.glob(os.path.join(test_dir, extension.upper())))
    
    if not image_paths:
        print("âŒ No images found in test directory")
        return
    
    print(f"ğŸ“ Loading test images from: {test_dir}")
    print(f"âœ… Found {len(image_paths)} test images")
    
    # Predictions dictionary
    predictions = {}
    all_confidences = []
    
    # Process each image
    print("ğŸ” Generating predictions...")
    with torch.no_grad():
        for image_path in tqdm(image_paths, desc="Predicting"):
            try:
                # Load and preprocess image
                image = Image.open(image_path).convert('RGB')
                input_tensor = transform(image).unsqueeze(0).to(device)
                
                # Make prediction
                logits = model(input_tensor)
                probabilities = torch.softmax(logits, dim=1)
                fake_prob = probabilities[0][1].item()  # Probability of being fake
                real_prob = probabilities[0][0].item()  # Probability of being real
                
                # Calculate confidence (max probability)
                confidence = max(fake_prob, real_prob)
                all_confidences.append(confidence)
                
                # Store prediction (filename without extension as key)
                filename = os.path.splitext(os.path.basename(image_path))[0]
                predictions[filename] = fake_prob
                
            except Exception as e:
                print(f"âŒ Error processing {image_path}: {e}")
                continue
    
    # Save predictions to JSON
    with open(output_json, 'w') as f:
        json.dump(predictions, f, indent=2)
    
    print(f"ğŸ’¾ Predictions saved to: {output_json}")
    
    # Print statistics
    fake_count = sum(1 for prob in predictions.values() if prob > 0.5)
    real_count = len(predictions) - fake_count
    
    # Calculate average confidence
    avg_confidence = np.mean(all_confidences) if all_confidences else 0
    
    print(f"ğŸ“Š Total predictions: {len(predictions)}")
    print(f"ğŸ“ˆ Prediction statistics:")
    print(f"   Real: {real_count} ({real_count/len(predictions)*100:.1f}%)")
    print(f"   Fake: {fake_count} ({fake_count/len(predictions)*100:.1f}%)")
    print(f"   Average confidence: {avg_confidence:.4f}")
    
    # Show first 10 predictions
    print(f"\nğŸ” First 10 predictions:")
    for i, (filename, prob) in enumerate(list(predictions.items())[:100]):
        pred_class = "fake" if prob > 0.5 else "real"
        confidence = max(prob, 1-prob)  # Confidence for the predicted class
        print(f"  {filename}: {pred_class} ({prob:.4f}, conf: {confidence:.4f})")
    
    if len(predictions) > 10:
        print(f"   ... and {len(predictions) - 10} more")
    
    print(f"\nğŸ¯ TESTING COMPLETED SUCCESSFULLY!")
    print(f"âœ… Success! Generated {len(predictions)} predictions")
    print(f"ğŸ“ Output file: {output_json}")
    
    return predictions

# ==================== USAGE ====================
if __name__ == "__main__":
    # UPDATE THESE PATHS
    MODEL_PATH = "C:/Users/nischay-ml-model/Documents/deepfake-images/vit_base_best.pth"
    TEST_DIR = "C:/Users/nischay-ml-model/Desktop/test3"  # or your test directory
    OUTPUT_JSON = "teamname_prediction.json"
    
    # Run prediction
    predictions = predict_test_images(
        model_path=MODEL_PATH,
        test_dir=TEST_DIR,
        output_json=OUTPUT_JSON
    )
```
